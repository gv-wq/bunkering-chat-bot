```python
import json
with open("./data/mobux_prices-example.json", "r") as fp:
    prices = json.load(fp)
```

```python
len(prices)
```

```python



from dotenv import load_dotenv
load_dotenv()


from app.services.internal_api.map_builder_api import MapBuilderApi
image_api = MapBuilderApi("http://127.0.0.1:8000")

from app.services.db_service import DbService
sql_db_service = DbService()
await sql_db_service.init_pool()
```

```python
ports = []
passed = []

for price in prices:
    country_name = price['port']["country"]['name']
    port_name = price["port"]['name']

    port_db, err = await sql_db_service.get_port_by_country_and_port_names(country_name, port_name)
    if port_db:
        ports.append(port_db)
    else:
        passed.append((country_name, port_name))

```
