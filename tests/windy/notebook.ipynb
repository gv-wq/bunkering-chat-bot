```python
import os
from dotenv import load_dotenv
load_dotenv()
import requests

windy_key = os.getenv("WINDY_KEY", None)
if windy_key is None:
    raise ValueError("WINDY_KEY environment variable not set")

base_url = "https://api.windy.com/api/point-forecast/v2"
```

```python
br1 = {
    "lat": 25.265354069777256,
    "lon": 55.219059880775326,
    "model": "gfsWave",
    "parameters": ["waves", "windWaves" ],
    "levels": ["surface"],
    "key": windy_key
}
```

```python
response = requests.post(base_url, json=br1)
j = response.json()
```

```python
j.keys()
```

```python
from datetime import datetime
r = [datetime.fromtimestamp(t/1000) for t in j.get("ts")]

min(r), max(r)
```

```python
import matplotlib.pyplot as plt

dt_seria = [datetime.fromtimestamp(t/1000) for t in j.get("ts")]
wave_heigth_seria = j.get("waves_height-surface")

fig, ax = plt.subplots(1, 1, figsize=(12, 4))
ax.plot(dt_seria, wave_heigth_seria)
ax.set_title("Высота волны возле Дубая")
ax.set_xlabel("Дата.")
ax.set_ylabel("Высота волны, м.")

plt.setp(ax.get_xticklabels(), rotation=45)
```
