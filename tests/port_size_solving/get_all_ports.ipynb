```python
import time
import asyncpg
from app.data.dto.main.SeaPort import SeaPortDB

db_name="BunkeringBot"
db_user="def_user"
db_password="super_password"
db_host="77.37.96.222"
db_port="5432"

connection_pool = await asyncpg.create_pool(
            database=db_name,
            user=db_user,
            password=db_password,
            host=db_host,
            port=db_port,
            min_size=1,
            max_size=20
        )
```

```python
ports = []

async with connection_pool.acquire() as conn:
    rows = await conn.fetch(
        """
        SELECT *
        FROM public.ports_vector_new
        """
    )

    for r in rows:
        p = SeaPortDB.from_db_row(r)
        ports.append(p)

len(ports)
```

```python
import json
with open("./data/all_ports.json", "w") as fp:

    fp.write(json.dumps([{
        "name": p.port_name,
        "locode": p.locode,
        "size": p.port_size,
        "latitude": p.latitude,
        "longitude": p.longitude,

    } for p in ports], indent=4))
```
