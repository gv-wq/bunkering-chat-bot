```python

from dotenv import load_dotenv
load_dotenv()

from app.services.db_service import DbService
sql_db_service = DbService()
await sql_db_service.init_pool()

import pandas as pd
df = pd.read_csv("./data/updated_port_sizes_28_01_2025.csv")
```

```python
update_df = df[df['Truck '].notna()].copy()
update_df['Truck '] = True
```

```python
u_list = {}
for _, r in update_df.iterrows():
    u_list[r['UN/LOCODE']] = {"truck_status": r["Truck "]}
```

```python
for k, v in u_list.items():
    await sql_db_service.update_port(k, v)
```
